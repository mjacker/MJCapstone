```python
%matplotlib inline
import pandas as pd
```

```python
df_canada = pd.read_csv("datasets/canadian_immegration_data.csv")
# adding index country to index
df_canada.index = df_canada['Country']

df_canada
```

```python
df_continents = df_canada.groupby("Continent", axis = 0).sum() # es como usar una base de datos y agrupar toda las suma de cada anho
df_continents
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

df_continents['Total'].head()
```

```python
df_continents['Total'].plot(kind='pie')
plt.title("Immigration to Canada by Continent [1980-2013]")
plt.show()
```
