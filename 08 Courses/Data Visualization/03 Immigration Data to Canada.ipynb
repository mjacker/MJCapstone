```python
import numpy as np
import pandas as pd
# from __future__ import print_function
```

```python
df_can = pd.read_csv("datasets/canadian_immegration_data.csv")
```

```python
df_can.head(5)
```

```python
# Creating Line Plots
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
years = list (map(str, range(1980, 2014)))
print(years)
```

```python
# adding index country to index
df_can.index = df_can['Country']

```

```python
# locating a specific country to get data
df_can.loc['Haiti', years] # only columns correspond to years

# the result is in the year 1980 there are 1666 immigrant to canada
```

```python
df_can.loc['Haiti', years].plot(kind = 'line')
plt.title("Immigration from Haiti")
plt.ylabel("Number of immigrants")
plt.xlabel("Years")
plt.show()
```

# Module 2
## Area Plots

```python
# sort dataframe by total
df_canada = pd.DataFrame.copy(df_can)
df_canada.drop('Country', inplace=True, axis=1)
df_canada.drop('Continent', inplace=True, axis=1)
df_canada.drop('Region', inplace=True, axis=1)
df_canada.drop('DevName', inplace=True, axis=1)

df_canada
```

```python
# df_test.sort_values('Country', axis=0)
df_top5 = pd.DataFrame.copy(df_canada.sort_values('Total', axis=0, ascending=False).head(5).transpose())
df_top5.drop('Total', inplace=True, axis=0)
df_top5
```

```python
import matplotlib as mpl

df_top5.plot(kind='area')
plt.title("Immigration trend of top 5 countries")
plt.ylabel("Number of immigrants")
plt.xlabel("Years")
plt.show()
```

## Histograms

```python
df_canada['2013'].plot(kind='hist')
plt.title('Histogram of Immigration from195 countries in 2013')
plt.ylabel("Number of Countries")
plt.xlabel("Number of Immigrants")
plt.show()
```

```python
# now try to do a better representation plot of the previous plot
import numpy as np
count, bin_edges = np.histogram(df_canada['2013'])
print (count)
print(bin_edges)
```

```python
df_canada['2013'].plot(kind='hist', xticks=bin_edges)
plt.title("Histogram of Immigration from 195 countries in 2013")
plt.ylabel("Number of countries")
plt.xlabel("Number of Immigrants")
plt.show()
```

## Bar Chart

```python
df_iceland = df_canada.loc["Iceland", years]
df_iceland
```

```python
df_iceland.plot(kind='bar')
plt.title("Icelandic immigrants to Canada from 1980 to 2013")
plt.xlabel("Year")
plt.ylabel("Number of immigrants")

# # Annotate arrow
plt.annotate('',  # s: str. will leave it blank for no text
             xy=(32, 70),  # place head of the arrow at point (year 2012 , pop 70)
             xytext=(28, 20),  # place base of the arrow at point (year 2008 , pop 20)
             xycoords='data',  # will use the coordinate system of the object being annotated
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='blue', lw=2)
             )

# Annotate Text
plt.annotate('2008 - 2011 Financial Crisis',  # text to display
             xy=(28, 29 ),  # start the text at at point (year 2008 , pop 30)
             rotation=73,  # based on trial and error to match the arrow
             va='bottom',  # want the text to be vertically 'bottom' aligned
             ha='left',  # want the text to be horizontally 'left' algned.
             )

plt.show()
```
