# Ten Thousand randoms numbers
1. By "standard normal" distribution.



```python
%matplotlib inline
from matplotlib.backends.backend_agg import  FigureCanvasAgg as FigureCanvas # agg stands for anti=grain geometry.
from matplotlib.figure import Figure
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creating Canvas and Figure base.
fig = Figure()
canvas = FigureCanvas(fig)
```

```python
# genrate random numbers
x = np.random.randn(10000)

ax = fig.add_subplot(111)
# 111 matlab convection, 1 row, 1 column, 1 grid for the location

ax.hist(x, bins=100)

# Decorate
ax.set_title("Normal distribution with $\mu=0, \sigma=1$")
fig.savefig("randomNumbers.png")
```

```python
# Drawing wit Scrpting Layer
plt.hist(x, 100)
plt.title("Normal distribution with $\mu=0, \sigma=1$")
plt.show()
```

```python
for _x in x:
    print(_x)
```

2. By 100 integer numbers


```python
array_numbers = np.random.randint(0, 100, 10000)
for _x in array_numbers:
    print(_x)
```

```python
# Drawing wit Scrpting Layer
plt.hist(array_numbers, 100)
plt.title("Random numbers distributions")
plt.show()
```
